```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np

x=np.linspace(-1000,1000,100)
y=np.linspace(-10,0,50)

X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='Reds')

plt.title("Reprsentation 3D de Z = x + y")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Cration des vecteurs x et y
x = np.linspace(-1000, 1000, 100)
y = np.linspace(-10, 0, 50)

# Cration de la grille 2D
X, Y = np.meshgrid(x, y)
# Calcul de Z sur toute la grille
Z = X**2 + Y**2

# Trac
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')  # Surface 3D colore

plt.title("Reprsentation 3D de Z = x + y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x,y):
    return x**2+y**2
x = np.linspace(-1000, 1000, 100)
y = np.linspace(-10, 0, 50)
x,y=np.meshgrid(x,y)
z=f(x,y)
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z, cmap='plasma')
plt.show()
```
